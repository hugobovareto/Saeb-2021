## (TG2Q18) Q049 - Pensando em seu desenvolvimento profissional, avalie se há necessidade de formação em cada uma das áreas listadas: Elaboração do projeto político pedagógico da instituição.

A questão 049 pode assumir os seguintes valores:
- **A:** nenhuma necessidade;
- **B:** pouca necessidade;
- **C:** muita necessidade.

```python
# Utilizar dataframe só com a questão de interesse:

saeb2021_prof_EI_Q049 = saeb2021_EI[['CO_UF', 'CO_MUNICIPIO',
                                'IN_CAPITAL',
                                'TP_LOCALIZACAO',
                                'ID_EDUCACAO_INFANTIL',
                                'TP_DEPENDENCIA_ADM_ESTRT',
                                'TP_EDUCACAO_INFANTIL',
                                'NU_DOMINIO',
                                'NU_ESTRATO',
                                'VALIDO_P',
                                'IN_ELEGIVEL',
                                'NAO_RESPOSTA',
                                'PESO_AMOSTRAL',
                                'TG2Q18']]
```

```python
saeb2021_prof_EI_Q049
```

```python
# Contagem de valores absolutos:
saeb2021_prof_EI_Q049['TG2Q18'].value_counts(dropna=False)
```

```python
# Contagem de valores ponderados: 
resultado1 = saeb2021_prof_EI_Q049.groupby('TG2Q18')['PESO_AMOSTRAL'].sum()

resultado1
```

```python
# Porcentagem dos valores:
porcentagens1 = (resultado1 / resultado1.sum()) * 100

round(porcentagens1, 2)
```

```python
# Criar um dicionário de correspondência entre valores reais e rótulos desejados:
correspondencia_rotulos = {
    'A': 'Nenhuma necessidade',
    'B': 'Pouca necessidade',
    'C': 'Muita necessidade'
}

# Substituir os valores reais pelos rótulos desejados
porcentagens1.index = porcentagens1.index.map(correspondencia_rotulos)
```

```python
# Visualização:
# Ajustar o tamanho da figura e a resolução
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)

# Gráfico de barras:
ax.bar(porcentagens1.index, porcentagens1, color='rebeccapurple')

# Adicionar rótulos de dados no topo das barras
for i, valor in enumerate(porcentagens1):
    ax.text(i, valor + 0.5, f'{valor:.2f}%', ha='center', va='bottom')

# Adicionar rótulos e título:
ax.set_ylabel('Porcentagem (%)')
ax.set_title('Pensando em seu desenvolvimento profissional, avalie se há necessidade de formação em cada uma das áreas listadas: \nElaboração do projeto político pedagógico da instituição.')

# Ajustar o valor final do eixo Y
ax.set_ylim(0, 60)  

# Salvar imagem:
plt.savefig('Saeb2021_EI_prof_Q049.png', bbox_inches='tight')

# Exibir o gráfico
plt.show()
```

## (EG2Q09) Q015 - Em uma semana normal de trabalho, quantas HORAS você costuma gastar, aproximadamente, com as seguintes atividades de direção da escola? Coordenar a gestão curricular, dos métodos de aprendizagem e da avaliação.

A questão 015 do diretor pode assumir os seguintes valores de 0h a 40h.

```python
# Utilizar dataframe só com a questão de interesse:

saeb2021_prof_EI_Q015 = saeb2021_EI[['CO_UF', 'CO_MUNICIPIO',
                                'IN_CAPITAL',
                                'TP_LOCALIZACAO',
                                'ID_EDUCACAO_INFANTIL',
                                'TP_DEPENDENCIA_ADM_ESTRT',
                                'TP_EDUCACAO_INFANTIL',
                                'NU_DOMINIO',
                                'NU_ESTRATO',
                                'VALIDO_P',
                                'IN_ELEGIVEL',
                                'NAO_RESPOSTA',
                                'PESO_AMOSTRAL',
                                'EG2Q09']]
```

```python
saeb2021_prof_EI_Q015
```

```python
# Contagem de valores absolutos:
saeb2021_prof_EI_Q015['EG2Q09'].value_counts(dropna=False)
```

```python
# Contagem de valores ponderados: 
resultado3 = saeb2021_prof_EI_Q015.groupby('EG2Q09')['PESO_AMOSTRAL'].sum()

resultado3
```

```python
# Crie um DataFrame a partir da Series:
df = pd.DataFrame({'index': resultado3.index, 'count': resultado3.values})

# Agrupe os valores do índice em intervalos de 5:
df['group'] = pd.cut(df['index'], bins=range(0, 41, 5), right=True, include_lowest=True)

# Agrupe por intervalo e calcule a soma das contagens
grouped_df = df.groupby('group')['count'].sum().reset_index()
```

```python
grouped_df
```

```python
# Porcentagem dos valores:
porcentagens3 = (grouped_df['count'] / grouped_df['count'].sum()) * 100

porcentagens3 = round(porcentagens3, 2)
porcentagens3.index.name = "EG2Q09"
porcentagens3
```

```python
# Criar um dicionário de correspondência entre valores reais e rótulos desejados:
correspondencia_rotulos = {
    0: '0h a 5h',
    1: '6h a 10h',
    2: '11h a 15h',
    3: '16h a 20h',
    4: '21h a 25h',
    5: '26h a 30h',
    6: '31h a 35h',
    7: '36h a 40h'
}

# Substituir os valores reais pelos rótulos desejados
porcentagens3.index = porcentagens3.index.map(correspondencia_rotulos)
```

```python
porcentagens3
```

```python
# Visualização:
# Ajustar o tamanho da figura e a resolução
fig, ax = plt.subplots(figsize=(6, 8), dpi=100)

# Gráfico de barras:
ax.bar(porcentagens3.index, porcentagens3, color='rebeccapurple')

# Adicionar rótulos de dados no topo das barras
for i, valor in enumerate(porcentagens3):
    ax.text(i, valor + 0.5, f'{valor:.2f}%', ha='center', va='bottom')

# Adicionar rótulos e título:
ax.set_ylabel('Porcentagem (%)')
ax.set_title('Em uma semana normal de trabalho, quantas HORAS você costuma gastar, \naproximadamente, com as seguintes atividades de direção da escola? \nCoordenar a gestão curricular, dos métodos de aprendizagem e da avaliação.') 
ax.set_xticklabels(list(porcentagens3.index), rotation=45, ha='right')

# Salvar imagem:
plt.savefig('Saeb2021_EI_diretor_Q015.png', bbox_inches='tight')

# Exibir o gráfico
plt.show()
```

## (TX_RESP_Q015) Q015 - Em uma semana normal de trabalho, quantas HORAS você costuma gastar, aproximadamente, com as seguintes atividades de direção da escola? Coordenar a gestão curricular, dos métodos de aprendizagem e da avaliação.

A questão 015 do diretor pode assumir os seguintes valores de 0h a 40h.

```python
# Utilizar dataframe só com a questão de interesse:

saeb2021_diretor_EF_Q015 = saeb2021_EF_diretor[['ID_REGIAO',
                                             'ID_UF',
                                             'ID_MUNICIPIO',
                                             'ID_AREA',
                                             'ID_ESCOLA',
                                             'IN_PUBLICA',
                                             'ID_LOCALIZACAO',
                                             'TX_RESP_Q015']]
```

```python
saeb2021_diretor_EF_Q015
```

```python
# Contagem de valores absolutos:
resultado12 = saeb2021_diretor_EF_Q015['TX_RESP_Q015'].value_counts(dropna=False)
resultado12
```

```python
# Crie um DataFrame a partir da Series:
df = pd.DataFrame({'index': resultado12.index, 'count': resultado12.values})

# Agrupe os valores do índice em intervalos de 5:
df['group'] = pd.cut(df['index'], bins=range(0, 41, 5), right=True, include_lowest=True)

# Agrupe por intervalo e calcule a soma das contagens
grouped_df = df.groupby('group')['count'].sum().reset_index()
```

```python
grouped_df
```

```python
# Porcentagem dos valores:
porcentagens12 = (grouped_df['count'] / grouped_df['count'].sum()) * 100

porcentagens12 = round(porcentagens12, 2)
porcentagens12.index.name = "TX_RESP_Q015"
porcentagens12
```

```python
# Criar um dicionário de correspondência entre valores reais e rótulos desejados:
correspondencia_rotulos = {
    0: '0h a 5h',
    1: '6h a 10h',
    2: '11h a 15h',
    3: '16h a 20h',
    4: '21h a 25h',
    5: '26h a 30h',
    6: '31h a 35h',
    7: '36h a 40h'
}

# Substituir os valores reais pelos rótulos desejados
porcentagens12.index = porcentagens12.index.map(correspondencia_rotulos)
```

```python
porcentagens12
```

```python
# Visualização:
# Ajustar o tamanho da figura e a resolução
fig, ax = plt.subplots(figsize=(6, 8), dpi=100)

# Gráfico de barras:
ax.bar(porcentagens12.index, porcentagens12, color='rebeccapurple')

# Adicionar rótulos de dados no topo das barras
for i, valor in enumerate(porcentagens12):
    ax.text(i, valor + 0.5, f'{valor:.2f}%', ha='center', va='bottom')

# Adicionar rótulos e título:
ax.set_ylabel('Porcentagem (%)')
ax.set_title('Em uma semana normal de trabalho, quantas HORAS você costuma gastar, \naproximadamente, com as seguintes atividades de direção da escola? \nCoordenar a gestão curricular, dos métodos de aprendizagem e da avaliação.') 
ax.set_xticklabels(list(porcentagens12.index), rotation=45, ha='right')

# Salvar imagem:
plt.savefig('Saeb2021_EF_diretor_TX_RESP_Q015.png', bbox_inches='tight')

# Exibir o gráfico
plt.show()
```
